```python
import numpy as np
import pandas as pd
import os
```

```python
#폴더 내 파일 불러오기
# 파일 위치 확인하기!!
files = os.listdir('./Raw_data')
print(len(files))
files
```

```python
def merge(files):
    global df
    df = pd.DataFrame()
    
    # csv파일 불러와서 데이터 프레임으로 저장
    for i in files:
        globals()['df_{}'.format(files.index(i))] = pd.read_csv(f'./Raw_data/{i}')
     
    # 컬럼명 통일
    for i in range(len(files)):
        if globals()['df_{}'.format(i)].columns[3] == 'lyric':
            globals()['df_{}'.format(i)].columns = ['x','가수','제목','가사']
            globals()['df_{}'.format(i)].drop(['x'], axis=1, inplace=True)
        
        
        elif globals()['df_{}'.format(i)].columns[3] == '노래 가사':
            globals()['df_{}'.format(i)].columns = ['x','가수','제목','가사']
            globals()['df_{}'.format(i)].drop(['x'], axis=1, inplace=True)
    
            
        elif globals()['df_{}'.format(i)].columns[3] == '가사':
            globals()['df_{}'.format(i)].columns = ['x','가수','제목','가사']
            globals()['df_{}'.format(i)].drop(['x'], axis=1, inplace=True)
    
        df = pd.concat([df,globals()['df_{}'.format(i)]],ignore_index=True)    
        
        
    # 결측치 제거    
    print('결측치 제거 전:',df.isna().sum(),'\n')
    df = df.dropna(subset=['가사'])
    print('결측치 제거 후:',df.isna().sum())
    
    
    # 가사 준비중인 컬럼 제거
    df = df[(df['가사'] != '가사준비중') & (df['가사'] != '[가사 준비중]')]
    
    # 중복값 제거
    df = df.drop_duplicates(['제목' , '가수'],keep = 'last')
    
    # 가수, 제목 제거
    df.pop('가수')
    df.pop('제목')
    
    # 가사 전처리
    df['가사']= df['가사'].str.replace(pat=r'[^\w]',repl=r' ',regex=True)
```

```python
merge(files)
df
```

```python
# 저장하기전에 파일명 한번 더 확인하기!!
df.to_csv("m0804.csv", encoding='utf-8-sig')
```
