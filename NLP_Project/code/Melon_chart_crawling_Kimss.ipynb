# '가수 이름' , '노래 제목' , '노래 가사' 크롤링

```python
## 크롬 윈도우 설정
# 크롬 옵션
options = webdriver.ChromeOptions()

# 크롬 윈도우 사이즈 조절
options.add_argument("--window-size=800,1200")
```

```python
# 중간에 멈췄을 때, 재시작 위치 계산해주는 함수
def location(x):
    y = x % 50
    z = (x-1) / 50
    
    return print(f'재시작 위치: {int(x - y + 1)}  중단된 페이지: {int(z+1)}p')

location(14238)
```

```python
# 크롤링한 결과를 저장할 리스트
singer_list = []
music_list = []
lyric_list = []

# 실행할 시간 랜덤값 지정
A = np.random.randint(1,4)

# # 크롤링 정지 방지용 휴식기간 변수
# relax = 1

## '가수 이름','노래 제목' 수집하는 반복문
# 크롤링 할 페이지 범위 설정 (위의 함수 사용)
for i in tqdm(range(46651,50000,50)):
    try:   
        chrome_path = chromedriver_autoinstaller.install()
        driver = webdriver.Chrome(chrome_path, options=options)
        driver.get(f"https://www.melon.com/genre/song_list.htm?gnrCode=GN0100#params%5BgnrCode%5D=GN0100&params%5BdtlGnrCode%5D=&params%5BorderBy%5D=POP&params%5BsteadyYn%5D=N&po=pageObj&startIndex={i}")
        
        # 크롤링 할 시간 간격 설정
        time.sleep(A)
        
        
        
        # 음원 랭킹
        rank_num = np.arange(1,51)
        
        for j in rank_num:
            # 곡 정보 들어가기
            driver.find_element_by_css_selector(f"tr:nth-child({j}) > td:nth-child(4) > div > a.btn.button_icons.type03.song_info").click()
            
            # 가수 이름 크롤링
            singer = driver.find_element_by_css_selector(".section_info .info .artist").text
            singer_list.append(singer)
            
            
            # 노래 제목 크롤링
            music = driver.find_element_by_css_selector(".section_info .info .song_name").text
            music_list.append(music)
            time.sleep(A)
            
            # 가사 o
            try:
                if driver.find_element_by_css_selector("button.button_more.arrow_d").text == '펼치기':
                    # 가사 펼치기
                    driver.find_element_by_css_selector(".section_lyric .wrap_lyric .button_more").click()
            
            
                    # 노래 가사 크롤링
                    lyric =  driver.find_element_by_css_selector("#d_video_summary").text #.replace('\n',' ')
                    lyric_list.append(lyric)
                    time.sleep(A)
            
            # 가사 x
            except:
                if driver.find_element_by_css_selector("#lyricArea > div > button").text == '가사등록하기':
                    lyric = "[가사 준비중]"
                    lyric_list.append(lyric)
                    time.sleep(A)

            j += 1
#             relax += 1
            
            #뒤로가기
            driver.back()
            time.sleep(A)

        
        # 페이지가 넘어감에 따라 +50
        rank_num = rank_num + 50
#         if relax % 50 == 0:
#             #크롤링한 페이지 입력 변수 저장
#             page_count = 100 + relax / 50
            
#             # 크롤링한 페이지 csv파일로 저장
#             melon_data = pd.DataFrame(list(zip(singer_list,music_list,lyric_list)), columns = ['가수','노래 제목','노래 가사'])
#             melon_data.to_csv(f"melon_ballad({page_count}p).csv", encoding='utf-8-sig')
            
#             #휴식
#             time.sleep(30)
    except:
        break
```

```python
print(music_list,'\n'*2,singer_list,'\n'*2,lyric_list)
```

```python
# 데이터 프레임화
melon_data = pd.DataFrame(list(zip(singer_list,music_list,lyric_list)), columns = ['가수','제목','가사'])
melon_data
```
